```python
import os, sys
import pandas as pd
import numpy as np
import pickle as pkl
```

```python
sys.path.append(os.path.abspath('../../GZ3D_production/'))
sys.path.append(os.path.abspath('../'))

import gz3d_fits
import spiral_resources
```

```python
available_paths = np.load('../available_v3_paths.npy')
```

```python
def load_galdict(path):
    with open(path, 'rb') as f:
        return pkl.load(f)
```

```python
def pc_pix_spiral(galdict, threshold = 5):
    data = gz3d_fits.gz3d_fits(galdict['filepath'])
    
    image_spiral_mask = data.spiral_mask
    pixels_above_threshold = (image_spiral_mask >= threshold).sum()
    
    return pixels_above_threshold * 100 / image_spiral_mask.size
```

```python
galdicts = []

i = 0

for root, directories, files in os.walk('../Form_Resources/Dicts/'):
    for name in files:
        path = os.path.join(root, name)
        galdict = load_galdict(path)
        
        galdict['pc_pix_spiral'] = pc_pix_spiral(galdict)
        
        galdicts.append(galdict)
        print(i)
        i += 1
```
