```python
import numpy as np
import os
```

```python
useful_paths = np.load('/raid5/homes/sshamsi/galaxy_zoo/GZ3D_spiral_analysis/usable_v3_paths.npy')
```

```python
useful_filenames = [path.split('/')[-1].split('.')[0] for path in useful_paths]
```

```python
def delete_files(dir_path):
    cov_list = []
    
    for root, dirs, files in os.walk(dir_path):
        for file in files:
            if file.endswith(".npy"):
                filename = file.split('.')[0]
                cov_list.append(filename)
                
    return cov_list
```

```python
for root, dirs, files in os.walk('cov_matrices/'):
    for file in files:
        if file.endswith(".npy"):
            filename = file[:-6]
            
            if filename in useful_filenames:
                print(file, ' was not removed!')
                continue
            else:
                os.remove(root + file)
                print(file, ' was removed!')
```
