### We'll go through all available spirals and then create dataframea and dictionaries for them for faster loading

```python
import numpy as np
import os, sys
import pandas as pd
import pathlib
import logging
```

```python
sys.path.append(os.path.abspath('../'))

import spiral_resources
import sfr_methods
from utils import save_dict, load_dict
```

```python
logging.basicConfig(filename='resource_run.log', encoding='utf-8', level=logging.DEBUG)
logger = logging.getLogger(__name__)
```

```python
available_spiral_paths = np.load('../Selecting_Sample/available_spirals.npy', allow_pickle=True)
```

```python
i = 0

for path in available_spiral_paths:
    filename = path.split('/')[-1]
    
    dictfile = pathlib.Path("Dicts/" + filename + '.pkl')
    dffile = pathlib.Path("DFs/" + filename + '.pkl')
    
    if (dictfile.exists() and dffile.exists()):
        print (i, filename, "already done.")
        i += 1
        continue
        
    elif dictfile.exists():
        galdict = load_dict('Dicts/' + filename + '.pkl')
        
        df = spiral_resources.return_df(galdict)
        df.to_pickle('DFs/' + filename + '.pkl')
        
    elif dffile.exists():
        galdict = spiral_resources.return_dict(path)
        save_dict(galdict, 'Dicts/' + filename + '.pkl')
        
    print(i)
    i += 1
```

We'll turn all the dictionaries into a comprehensive DataFrame as this'll enable us to use it faster.

```python
galdict_list = []

for root, directories, files in os.walk('Dicts/'):
    for name in files:
        path = 'Dicts/' + name
        galdict = load_dict(path)
        galdict_list.append(galdict)
```

```python
main_df = pd.DataFrame.from_dict(galdict_list)
```

### Now we add different thresholds to each DataFrame

```python
thresholds = []

for i in range(1, 8):
    for j in range(1, 8):
        thresholds.append((i, j))
```

```python
k = 0

for root, directories, files in os.walk('Dicts/'):
    for name in files:        
        galdict = load_dict('Dicts/' + name)
        galdf = pd.read_pickle('DFs/' + name)
        
        for threshold in thresholds:
            colsp = 'sp_{Tsp}{Tnsp}'.format(Tsp=threshold[0], Tnsp=threshold[1])
            colnsp = 'nsp_{Tsp}{Tnsp}'.format(Tsp=threshold[0], Tnsp=threshold[1])
            
            if not {colsp, colnsp}.issubset(galdf.columns):
                galdf = spiral_resources.update_spirals(galdf, galdict['filepath'], galdict['map_shape'], spiral_threshold=threshold[0],
                                                        other_threshold=threshold[1])
                galdf.to_pickle('DFs/' + name)
                
        print(k)
        k += 1
```
