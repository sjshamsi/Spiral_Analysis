```python
import numpy as np
import os, sys
import pandas as pd
import pickle
import pathlib
import logging
```

```python
def save_dict(obj, name):
    with open('Dicts/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_dict(name):
    with open('Dicts/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
sys.path.append(os.path.abspath('../'))

import spiral_resources
import sfr_methods
```

```python
logging.basicConfig(filename='resource_run.log', encoding='utf-8', level=logging.DEBUG)
logger = logging.getLogger(__name__)
```

```python
available_paths = np.load('../Select_Threshold/available_filepaths.npy', allow_pickle=True)
```

```python
i = 0

for path in available_paths:
    filename = path.split('/')[-1]
    
    dictfile = pathlib.Path("Dicts/" + filename + '.pkl')
    dffile = pathlib.Path("DFs/" + filename + '.pkl')
    
    if (dictfile.exists() and dffile.exists()):
        print (i, filename, "already done.")
        i += 1
        
    else:
        try:
            galdict = spiral_resources.return_dict(path)
            save_dict(galdict, filename)

            df = spiral_resources.return_df(galdict)
            df.to_pickle('DFs/' + filename + '.pkl')

        except Exception:
            logger.error("Fatal error in forming resources for {}".format(filename), exc_info=True)
            
        print(i)
        i += 1
```

### Now we add different thresholds to each DataFrame

```python
thresholds = []

for i in range(1, 8):
    for j in range(1, 8):
        thresholds.append((i, j))
```

```python
k = 0

for root, directories, files in os.walk('Dicts/'):
    for name in files:
        filename = '.'.join(name.split('.')[0:-1])
        
        galdict = load_dict(filename )
        galdf = pd.read_pickle('DFs/' + filename + '.pkl')
        
        for threshold in thresholds:
            colsp = 'sp_{Tsp}{Tnsp}'.format(Tsp=threshold[0], Tnsp=threshold[1])
            colnsp = 'nsp_{Tsp}{Tnsp}'.format(Tsp=threshold[0], Tnsp=threshold[1])
            
            if not {colsp, colnsp}.issubset(galdf.columns):
                galdf = spiral_resources.update_spirals(galdf, galdict['filepath'], galdict['map_shape'], spiral_threshold=threshold[0],
                                                        other_threshold=threshold[1])
                galdf.to_pickle('DFs/' + filename + '.pkl')
                
        print(k)
        k += 1
```
