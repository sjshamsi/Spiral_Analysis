```python
plt.style.use('seaborn')
plt.style.use('seaborn-colorblind')
```

```python
usable_paths = np.load('../usable_v3_paths.npy')
```

```python
def integrated_sfr(galdf, galdict, avg, threshold):
    df = galdf.copy()
    
    df = df.dropna()
    df = df[(df['$r/r_e$'] < 1.5) & (df['$r/r_e$'] > 0.1)]
    df = df[(df['Comp'] == 0) & (df['AGN'] == 0) & (df['Seyfert'] == 0) & (df['Liner'] == 0)]
    df = df[(df['Spiral Arm ' + str(threshold)] == True) | (df['Nonspiral Arm ' + str(threshold)] == True)]
    
    spiral_df = df[(df['Spiral Arm ' + str(threshold)] == True) & (df['Nonspiral Arm ' + str(threshold)] == False)]
    nspiral_df = df[df['Nonspiral Arm ' + str(threshold)] == True]
    
    spiral_sfr, sig_spiral_sfr = sfr_methods.get_sfr(spiral_df.index, galdf, galdict, avg=avg)
    nspiral_sfr, sig_nspiral_sfr = sfr_methods.get_sfr(nspiral_df.index, galdf, galdict, avg=avg)
    combined_sfr, sig_combined_sfr = sfr_methods.get_sfr(df.index, galdf, galdict, avg=avg)
    
    return spiral_sfr, sig_spiral_sfr, nspiral_sfr, sig_nspiral_sfr, combined_sfr, sig_combined_sfr
```

```python
def load_dict(path):
    with open(path, 'rb') as f:
        return pickle.load(f)
```

```python
def calculate_stdv(series):
    mean = series.mean()
    
    summ = 0
    for element in series:
        summ = summ + (element - mean)**2
    
    return np.sqrt(summ / len(series))
```

```python
dict_list = []
df_list = []

for i in range(len(usable_paths)):
    filename = usable_paths[i].split('/')[-1].split('.')[0]
    
    galdict = load_dict('../Form_Resources/Dicts/' + filename + '.dict.pkl')
    df = pd.read_pickle('../Form_Resources/DFs/' + filename + '.df.pkl')
    
    dict_list.append(galdict)
    df_list.append(df)
```

```python
thresholds = []

for i in range(1, 8):
    for j in range(1, 8):
        thresholds.append((i, j))
```

```python
%%time

for i in range(len(usable_paths)):
    galdict = dict_list[i]
    df = df_list[i]
    
    k = 0
    for threshold in thresholds:
        spiral_sfr, sig_spiral_sfr, nspiral_sfr, sig_nspiral_sfr, combined_sfr, sig_combined_sfr = integrated_sfr(df, galdict, False, threshold)
        
        galdict['tot_spiral_sfr ' + str(threshold)] = spiral_sfr
        galdict['sig_tot_spiral_sfr ' + str(threshold)] = sig_spiral_sfr
        
        galdict['tot_nonspiral_sfr ' + str(threshold)] = nspiral_sfr
        galdict['sig_tot_nonspiral_sfr ' + str(threshold)] = sig_nspiral_sfr
        
        galdict['tot_combined_sfr ' + str(threshold)] = combined_sfr
        galdict['sig_tot_combined_sfr ' + str(threshold)] = sig_combined_sfr
        
        print(k)
        k += 1
        
    print('Galaxy #', i)
```

```python
def save_dict(galdict, name):
    with open(name + '.dict.pkl', 'wb') as f:
        pickle.dump(galdict, f, pickle.HIGHEST_PROTOCOL)
```

```python
for galdict in dict_list:
    save_dict(galdict, '../Form_Resources/Dicts_Total_SFR_Thresholds/' + galdict['filename'].split('.')[0])
```
